```python
import matplotlib.pyplot as plt

# Dados para o gráfico de pizza
labels = ['Categoria A', 'Categoria B', 'Categoria C', 'Categoria D']
sizes = [25, 30, 15, 30]

# Configurar cores
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99']

# Criar o gráfico de pizza
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=colors)

# Adicionar título
plt.title('Distribuição de Categorias')

# Exibir o gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt

# Dados
labels = ['Categoria A', 'Categoria B', 'Categoria C']
sizes = [40, 30, 30]
explode = (0.1, 0, 0)  # Destacar a primeira fatia

# Criar o gráfico de rosca com destaque
fig, ax = plt.subplots()
wedges, text, autotext = ax.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, explode=explode, wedgeprops=dict(width=0.2))

# Adicionar círculo no centro para criar o efeito de rosca
centre_circle = plt.Circle((0, 0), 0.1, color='white', linewidth=0.5)
ax.add_patch(centre_circle)

# Adicionar título
ax.set_title('Gráfico de Rosca com Destaque')

# Exibir o gráfico
plt.show()
```
