## Análise Comparativa de Métricas de Desempenho dos Modelos de AM - Gráficos de Barra, Gráficos de Linha, Gráficos de Radar

```python
import matplotlib.pyplot as plt
import pandas as pd
from io import StringIO

# Seus dados em formato CSV
dados_csv = """
Modelo;Acurácia;Precisão;Recall;F1-Score
Local Outlier Factor;83.87;96.30;86.67;91.23
Isolation Forest;58.06;100;56.67;72.34
DBScan;93.55;87.50;87.50;87.50
"""

# Criar um DataFrame a partir dos dados CSV
df = pd.read_csv(StringIO(dados_csv), sep=';')

# Configurar cores na tonalidade azul
cores = ['#98c2ff', '#aec7e8', '#1f77b4', '#8da0cb']

# Configurar o gráfico
fig, ax = plt.subplots(figsize=(6, 3))

# Iterar sobre as colunas e criar barras separadas para cada métrica
width = 0.2  # Largura das barras
models = df['Modelo'].tolist()
metrics = ['Acurácia', 'Precisão', 'Recall', 'F1-Score']

for i, metric in enumerate(metrics):
    ax.bar([pos + width * i for pos in range(len(models))], df[metric], width=width, label=metric, color=cores[i])

# Adicionar legendas e rótulos
ax.set_xlabel('Modelo')
ax.set_ylabel('Percentagem')
# ax.set_title('Desempenho dos Modelos por Métrica')
ax.set_xticks([pos + width for pos in range(len(models))])
ax.set_xticklabels(models)
ax.legend(loc='lower left', bbox_to_anchor=(1, 0.58))

# Exibir o gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from io import StringIO

# Seus dados em formato CSV
dados_csv = """
Modelo;Acurácia;Precisão;Recall;F1-Score
Local Outlier Factor;83.87;96.30;86.67;91.23
Isolation Forest;58.06;100;56.67;72.34
DBScan;93.55;87.50;87.50;87.50
"""

# Criar um DataFrame a partir dos dados CSV
df = pd.read_csv(StringIO(dados_csv), sep=';')

# Extrair nomes dos modelos e métricas
modelos = df['Modelo'].tolist()
metricas = df.columns[1:]

# Número de categorias (métricas)
num_metricas = len(metricas)

# Criar ângulos para as métricas
angulos = np.linspace(0, 2 * np.pi, num_metricas, endpoint=False).tolist()

# Unir o ponto inicial e final para criar um gráfico fechado
valores_modelos = df.iloc[:, 1:].values / 100.0  # Normalizar os valores para o intervalo [0, 1]
valores_modelos = np.concatenate((valores_modelos, valores_modelos[:, 0:1]), axis=1)
angulos += angulos[:1]

# Configurar o gráfico de radar
fig, ax = plt.subplots(figsize=(4, 4), subplot_kw=dict(polar=True))
ax.fill(angulos, valores_modelos[0], 'b', alpha=0.1, label=modelos[0])
ax.fill(angulos, valores_modelos[1], 'r', alpha=0.1, label=modelos[1])
ax.fill(angulos, valores_modelos[2], 'g', alpha=0.1, label=modelos[2])

# Adicionar linhas conectando os pontos
ax.set_thetagrids(np.degrees(angulos[:-1]), metricas)
ax.plot(angulos, valores_modelos[0], 'b', linewidth=2)
ax.plot(angulos, valores_modelos[1], 'r', linewidth=2)
ax.plot(angulos, valores_modelos[2], 'g', linewidth=2)

# Ajustar a posição da legenda
ax.legend(loc='upper right', bbox_to_anchor=(1.6, 1.1))

# Adicionar título
# ax.set_title('Desempenho dos Modelos por Métrica')

# Exibir o gráfico de radar
plt.show()
```

```python
from plotnine import ggplot, aes, geom_line, theme, theme_minimal, ylim, labs

# Seus dados em formato CSV
dados_csv = """
Modelo;Acurácia;Precisão;Recall;F1-Score
Local Outlier Factor;83.87;96.30;86.67;91.23
Isolation Forest;58.06;100;56.67;72.34
DBScan;93.55;87.50;87.50;87.50
"""

# Criar um DataFrame a partir dos dados CSV
df = pd.read_csv(StringIO(dados_csv), sep=';')

# Mapear os dados para o formato longo
df_long = pd.melt(df, id_vars=['Modelo'], var_name='Métrica', value_name='Valor')

# Criar o gráfico de radar com plotnine
plot = (ggplot(df_long, aes(x='Métrica', y='Valor', group='Modelo', color='Modelo')) +
        geom_line() +
        theme_minimal() +
        theme(legend_position='top') +
        ylim(50, 100) +
        labs(#title='Desempenho dos Modelos por Métrica',
             x='Métrica',
             y='Percentagem'))

# Exibir o gráfico de radar
print(plot)
```
